```python
import matplotlib.pyplot as plt
import wave
import numpy as np
```

```python
fileat=wave.open('fileghiam.wav','r')
```

```python
Fs=fileat.getframerate()
print('tần số lấy mẫu Fs là : ' + str(Fs))
```

```python
sw=fileat.getsampwidth()
bits=8*sw
print('số bits biểu diễn là : ' + str(bits))
```

```python
sokenh= fileat.getnchannels()
print('số kênh âm thanh là : ' + str(sokenh))
```

```python
tinhieu = fileat.readframes(-1)
tinhieu = np.fromstring(tinhieu, 'Int16')

tgian = np.linspace(0, len(tinhieu)/Fs,num=len(tinhieu))
```

```python
plt.figure(1)
plt.title('đồ thị hiển thị tín hiệu âm thanh')
plt.plot(tgian,tinhieu)
plt.xlabel('trục thời gian t(s)')
```
